```python
import pandas as pd
```

```python
data = pd.read_csv('Concrete_Data.csv')
data = data[data['Age'] == 28]
print(type(data))
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
print(x)
print(y)
```

```python
from matplotlib import pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x,y)
print(reg.score(x,y))
print(reg.coef_)
print(reg.intercept_)

yp= reg.predict(x)
plt.plot(y)
plt.plot(yp)
plt.show()
```
