## Data visualisation using Matplotlib

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
y = [5,3,4,5,6,7,2,8,9]
plt.plot(y)
plt.show()
```

```python
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
```

```python
print(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin')
plt.title('sin wave')
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('angle')
plt.ylabel('sin')
plt.title('sin wave')
plt.show()
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel('angle')
plt.ylabel('sin')
plt.title('sin wave')
plt.show()
```

```python
plt.plot(x,np.sin(x), 'r--', label='sin')
plt.plot(x,np.cos(x) , 'g--', label='cos') #Scatter connections
plt.xlabel('angle')
plt.ylabel('sin/cos')
plt.title('sin/cos wave')
plt.ylim(-2,2)
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = np.array([2,3,5,6,7,8,9,0,5,4])
plt.plot(x,y)
plt.show()
```

## Subplot

```python
fig, axis = plt.subplots(1,2, figsize=(10,5))
x = np.arange(0,10,0.1)
axis[0].plot(x,np.sin(x), 'g--')
axis[0].set_title('sin')
axis[1].plot(x,np.cos(x), 'r--')
axis[1].set_title('cos')
plt.show()
```

```python
fig, axis = plt.subplots(2,2, figsize=(10,10))
x = np.arange(0,10,0.1)
axis[0][1].plot(x,np.sin(x), 'y--')
axis[0][1].set_title('sin')
axis[1][0].plot(x,np.sin(x), 'g--')
axis[1][0].set_title('sin')
axis[1][1].plot(x,np.cos(x), 'b--')
axis[1][1].set_title('cos')
axis[0][0].plot(x,np.cos(x), 'r--')
axis[0][0].set_title('cos')
plt.show()
```
